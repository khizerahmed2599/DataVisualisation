### Correlation Matrix

```python
corr = df.corr()
corr
```

```python
sns.set(rc = {"figure.figsize":(17,10)})
sns.set_theme(style="white")
sns.plotting_context()
cmap = sns.diverging_palette(240, 10, n=9)
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, annot = True, cmap=cmap, mask=mask)
```

```python
data = df[['artists', 'popularity','name', 'year']]
```

```python
data.head()
```

```python
data.shape
```

```python
data['artists'] = data['artists'].str[1:-1]
data.head()
```

### Getting Individual Artists

```python
d=data.assign(artists=data.artists.str.split(",")).explode('artists')
```

```python
d.head()
```

```python
d.shape
```

```python
d.tail(10)
```

```python
#d['artists'].value_counts()
len(np.unique(d['artists']))
```

```python
d.isnull().sum()
```

```python
d.duplicated().sum()
```

```python
d = d.reset_index(drop=True)
```

```python
d.head()
```

```python
len(np.unique(d['artists']))
```

### Determing the Average Popularity of Each Artist in Each year

```python
d1 = d.groupby(['artists','year']).agg({'popularity':[np.mean]}).reset_index()
```

```python
d1
```

### Saving The Final dataset

```python
d1.to_csv('d1.csv')
```
